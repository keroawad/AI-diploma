```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
date=input("enter a datat for scrapping:yyyy-mm-dd").strip()
page=requests.get(f'https://www.filgoal.com/matches/?date={date}')
page
```

```python
src=page.text
src
soup=BeautifulSoup(src,'html')
soup
```

```python
csvfile=open("filgoal.csv","w",encoding="utf-8")
writer=csv.writer(csvfile)
```

```python
writer.writerow(["champo","team1","team2","stat"])
```

```python
champion=soup.find_all("div",class_="mc-block")
champion
for i in champion:
    champo = i.find("h6").text.strip()
    team_a = i.find("div", class_="f").find("strong").text.strip()
    team_b = i.find("div", class_="s").text.strip()
    stat = i.find("div", class_="m").text.strip()
    print(champo,team_a,team_b,stat)
    writer.writerow([champo,team_a,team_b,stat])
csvfile.close()    
```
