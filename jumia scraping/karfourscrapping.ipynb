```python
csvfile=open("jomai.csv","w",encoding="utf-8")
writer=csv.writer(csvfile)
```

```python
writer.writerow(["model","price","Evaluate"])
```

```python
for i in range(3):
    url = f"https://www.jumia.com.eg/ar/catalog/?q=%D9%83%D9%88%D8%AA%D8%B4%D9%8A&page={i}"
    website = requests.get(url).text
    soup = BeautifulSoup(website, "html.parser")
    models = soup.find_all("article", class_="prd _fb col c-prd")
    for mod in models:
        model = mod.find("div", class_="info").find("h3", class_="name").text.strip()
        price = mod.find("div", class_="info").find("div", class_="prc").text.strip()
        if mod.find("div", class_="info").find("div", class_="rev"):
            Evaluate = mod.find("div", class_="info").find("div", class_="rev").find("div", class_="stars _s").text.strip()
        else:
            Evaluate = "No rating"
        print(model, price, Evaluate)
        writer.writerow([model, price, Evaluate])
csvfile.close()
```
